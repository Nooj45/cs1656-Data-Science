```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import dateutil

filter_month = 4
filter_stationID = 1046

df1 = pd.read_csv('HealthyRideRentals2019-Q1.csv')
df2 = pd.read_csv('HealthyRideRentals2019-Q2.csv')
df3 = pd.read_csv('HealthyRideRentals2019-Q3.csv')
df = pd.concat([df1, df2, df3])
df_small = df[['Starttime', 'Stoptime', 'From station id', 'To station id']].copy()

#getting month format
df_small['Starttime'] = pd.to_datetime(df_small['Starttime'])
df_small['Starttime'] = df_small['Starttime'].map(lambda x: x.date())
df_small['Stoptime'] = pd.to_datetime(df_small['Stoptime'])
df_small['Stoptime'] = df_small['Stoptime'].map(lambda x: x.date())

month = df_small['Starttime'].map(lambda x: x.month) == filter_month
df_small = df_small[month]
df_small = df_small.loc[df_small['From station id'] == filter_stationID]

if df_small['From station id'] is not None:
    fromCNT = pd.crosstab(df_small['From station id'], [df_small['Starttime'], df_small['Stoptime']], margins=True, margins_name='Total')

label=[]
for x in range(0, 31):
    label.append(x + 1)

# having problems with the graph
y = fromCNT['Total'].head(30)    
y_pos = np.arange(len(label))
plt.bar(y_pos, height=y)
plt.xticks(y_pos, label)
plt.xlabel('Days of the Month')
plt.ylabel('fromCNT')
plt.title('Task 2.2 Distribution of Bike Rentals During the Month')
plt.rcParams['figure.figsize'] = (10,6)
plt.show()
```

### Task 3.1

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df1 = pd.read_csv('HealthyRideRentals2019-Q1.csv')
df2 = pd.read_csv('HealthyRideRentals2019-Q2.csv')
df3 = pd.read_csv('HealthyRideRentals2019-Q3.csv')
df = pd.concat([df1, df2, df3])

df_small = df[['Starttime', 'Stoptime', 'From station id', 'To station id']]
if df_small['From station id'] is not None:
    fromCNT = pd.crosstab(df_small['From station id'], [df_small['Starttime'], df_small['Stoptime']], margins=True, margins_name='Total')
    
if df_small['To station id'] is not None:
    toCNT = pd.crosstab(df_small['To station id'], [df_small['Starttime'], df_small['Stoptime']], margins=True, margins_name='Total')
    
rebalCNT = fromCNT['Total'].subtract(toCNT['Total'], fill_value=0)

y = rebalCNT.head(25)
xlab = df_small['From station id'].head(25)

plt.bar(x=np.arange(1,26), height=y)
plt.xticks(np.arange(1,26), xlab)
plt.xlabel('Station id')
plt.ylabel('rebalCNT per station')
plt.title('Task 3.1 TOP 25 Stations w/ rebalCNT')
plt.rcParams['figure.figsize'] = (15, 11)
plt.show()
```

### Task 3.2

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import dateutil

filter_month = 4
filter_stationID = 1046

df1 = pd.read_csv('HealthyRideRentals2019-Q1.csv')
df2 = pd.read_csv('HealthyRideRentals2019-Q2.csv')
df3 = pd.read_csv('HealthyRideRentals2019-Q3.csv')
df = pd.concat([df1, df2, df3])
df_small = df[['Starttime', 'Stoptime', 'From station id', 'To station id']].copy()

#getting month format
df_small['Starttime'] = pd.to_datetime(df_small['Starttime'])
df_small['Starttime'] = df_small['Starttime'].map(lambda x: x.date())
df_small['Stoptime'] = pd.to_datetime(df_small['Stoptime'])
df_small['Stoptime'] = df_small['Stoptime'].map(lambda x: x.date())

month = df_small['Starttime'].map(lambda x: x.month) == filter_month
df_small = df_small[month]
df_small = df_small.loc[df_small['From station id'] == filter_stationID]

if df_small['From station id'] is not None:
    fromCNT = pd.crosstab(df_small['From station id'], [df_small['Starttime'], df_small['Stoptime']], margins=True, margins_name='Total')
    
if df_small['To station id'] is not None:
    toCNT = pd.crosstab(df_small['To station id'], [df_small['Starttime'], df_small['Stoptime']], margins=True, margins_name='Total')
    
rebalCNT = fromCNT['Total'].subtract(toCNT['Total'], fill_value=0)

y = rebalCNT.head(30)
label=[]
for x in range(0, 31):
    label.append(x + 1)
    
# having problems with plotting
y_pos = np.arange(len(label))
plt.bar(y_pos, height=y)
plt.xticks(y_pos, label)
plt.xlabel('Days of the Month')
plt.ylabel('rebalCNT')
plt.title('Task 2.2 Distribution of Bike Rebalancing During the Month')
plt.rcParams['figure.figsize'] = (10,6)
plt.show()
```
